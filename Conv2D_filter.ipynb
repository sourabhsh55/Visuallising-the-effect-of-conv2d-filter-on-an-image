```
from keras.models import Sequential
import keras
import os
import numpy as np
from keras.layers import Dense, Conv2D, MaxPooling2D
import cv2
import matplotlib.pyplot as plt
from sklearn import datasets
```

```
items = os.listdir(os.getcwd())
items
```

```
shinchan = cv2.imread(items[1])
# print(shinchan.shape) # coloured image shape.
print(shinchan[:,:,1].shape)
# plt.imshow(shinchan)
plt.imshow(shinchan[:,])
print(shinchan[:,:,0].shape)
```

```
print(shinchan.shape)
# np.ravel(shinchan).shape
# 224436/3
317*236
```

```
model = Sequential()
```

```
c1 = Conv2D(filters=1,kernel_size=(3,3),strides=(1, 1),padding='SAME',input_shape=shinchan.shape)
c2 = Conv2D(filters=1,kernel_size=(3,3),strides=(1, 1),padding='SAME')
p1 = MaxPooling2D(pool_size=(2,2),strides=None,padding='SAME')
p2 = MaxPooling2D(pool_size=(2,2),strides=None,padding='SAME')
c3 = Conv2D(filters=1,kernel_size=(3,3),strides=(1, 1),padding='SAME')
c4 = Conv2D(filters=1,kernel_size=(3,3),strides=(1, 1),padding='SAME')
p3 = MaxPooling2D(pool_size=(2,2),strides=None,padding='SAME')
p4 = MaxPooling2D(pool_size=(2,2),strides=None,padding='SAME')
```

```
model.add(c1)
# model.add(p1)
model.add(c2)
# model.add(p2)
model.add(c3)
model.add(p3)
model.add(c4)
model.add(p4)
```

```
con_batch=np.expand_dims(shinchan,axis=0)
con_batch.shape
```

```
con_batch_out= model.predict(con_batch)
con_batch_out.shape
```

```
image = np.squeeze(con_batch_out,axis=0)
print(image.shape)
image = image.reshape(image.shape[:2])
print(image.shape)
print(image.shape[:2])
```
